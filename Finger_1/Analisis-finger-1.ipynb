```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
events = pd.read_csv('../dataTP1/events.csv', low_memory=False)
```

```python
events.columns
```

```python
events.head(12)
```

```python
tipos_de_evento = events.groupby('event').count()
```

```python
tipos_de_evento.head(10)
```

```python
eventos_mas_repetidos = events['event'].value_counts().head(10).to_frame('cantidad')
eventos_mas_repetidos
```

```python
#otra forma
events.groupby('event').agg('size').sort_values(ascending=False).head(3)
```

```python
sns.set(style="whitegrid")
sns.set(rc={'figure.figsize':(13,8.27)})
plot_eventos = sns.barplot(x=eventos_mas_repetidos.index, y=eventos_mas_repetidos['cantidad'])
plot_eventos.set(title='Eventos y sus cantidades', xlabel='Tipo de evento', ylabel='Cantidad de eventos')
```

```python
modelos_mas_convertidos = events.loc[events.event.str.contains('conversion'),['event','sku']]
```

```python
conversiones = events.loc[events.event.str.contains('conversion'),['event','sku','model']]
```

```python
modelos_mas_convertidos = conversiones.groupby(['sku','model']).agg('size').to_frame('cantidad')
```

```python
modelos_mas_convertidos = modelos_mas_convertidos.reset_index().sort_values('cantidad',ascending=False).head(5)
```

```python
modelos_mas_convertidos
```

```python
sns.set(style="whitegrid")
sns.set(rc={'figure.figsize':(11.7,8.27)})
plot_conversiones = sns.barplot(order=modelos_mas_convertidos['sku'], x=modelos_mas_convertidos['sku'], y=modelos_mas_convertidos['cantidad'].head(5))#, order=modelos_mas_convertidos['cantidad'])                             
plot_conversiones.set_title('Modelos más convertidos')
plot_conversiones.set(xlabel='Modelo', ylabel='Cant. de conversiones')
#plot_conversiones.order(modelos_mas_convertidos['cantidad'])
```

```python
actividad_por_personas = events['person'].value_counts().head(10).to_frame('cantidad')
```

```python
actividad_por_personas
```

```python
sns.set(style="whitegrid")
sns.set(rc={'figure.figsize':(11.7,8.27)})
plot_personas = sns.barplot(x=actividad_por_personas.index, y=actividad_por_personas['cantidad'])
plot_personas.set_title('Personas más activas')
plot_personas.set(xlabel='Id Persona', ylabel='Cant. de actividades')
```
